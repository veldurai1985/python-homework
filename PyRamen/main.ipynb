```python
# @TODO: Import libraries
import csv
from pathlib import Path

# Check the current directory where the Python program is executing from
print(f"Current Working Directory: {Path.cwd()}")

# @TODO: Set file paths for menu_data.csv and sales_data.csv
menu_filepath = Path('./Resources/menu_data.csv')
sales_filepath = Path('./Resources/sales_data.csv')

# @TODO: Initialize list objects to hold our menu and sales data
menu = []
sales = []

#Initialize all the variables


# @TODO: Read in the menu data into the menu list
with open(menu_filepath, 'r') as menu_data_file:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    menu_data = csv.reader(menu_data_file, delimiter=',')

    # Storing the header data from the start of the file
    menu_header = next(menu_data)
    
    # Iterating over the menu data and loading into menu list 
    for row in menu_data: 
        # Creating menu list 
        menu.append(row)

# @TODO: Read in the sales data into the sales list
with open(sales_filepath, 'r') as sales_data_file:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    sales_data = csv.reader(sales_data_file, delimiter=',')

    # Storing the header data from the start of the file
    sales_header = next(sales_data)
    
    # Iterating over sales data and loading into sales list 
    for row in sales_data: 
        # Creating sales list
        sales.append(row)


# @TODO: Initialize dict object to hold our key-value pairs of items and metrics
report = {}

# Creating a menu list and menu_list_unique to identify the unique menu item
menu_list = []
menu_list_unique = []

# Initialize a row counter variable
sales_row_count = 0

# @TODO: Loop over every row in the sales list object

for row in sales:
    menu_item = row[4]
    menu_list.append(menu_item)
    sales_row_count += 1

# Loop over menu_list to identify the unique menu item based on sales data
for unique_item in menu_list: 
    # check if exists in unique_list or not 
    if unique_item not in menu_list_unique: 
        menu_list_unique.append(unique_item) 

# Creating the intialized report dictionary for the menu item in sales data
for menu_dictionary in menu_list_unique: 
    # check if exists in unique_list or not 
    report[menu_dictionary] = {"01-count": 0, "02-revenue": 0, "03-cogs": 0, "04-profit": 0} 


# Main part
# Looping through the menu item and collecting metrics for that menu item 

for data in menu:
    # Getting the menu item and coverting the price and cost to float 
    menu_item_2 = data[0]
    price = float(data[3])
    cost = float(data[4])
    # Profit is calculated for each menu item 
    profit = price - cost

    # Initialize values for sales data looping 
    item_quantity = 0

    # Looping through sales item for each menu item
        
    for element in sales:
        # Getting the sales item and coverting the quantity to int 
        sales_item_2 = element[4]
        quantity = int(element[3])
        # Determining if the sales item is equal to menu item
        if menu_item_2 == sales_item_2:
            # if the sales item is matched aggreating the item quantity
            item_quantity  += quantity
        else: 
            print (f"{sales_item_2} does not equal {menu_item_2}! NO MATCH!")
    
    # Writing the metrics to the report 
    if menu_item_2 in report:
        report[menu_item_2]["01-count"] = item_quantity
        report[menu_item_2]["02-revenue"] = price * item_quantity
        report[menu_item_2]["03-cogs"] = cost * item_quantity
        report[menu_item_2]["04-profit"] = profit * item_quantity

#Print the number of rows in sales data
print(f"Number of records in sales data: {sales_row_count}")

# Set the output file path
output_path = Path('pyramen_output.csv')

# Open the output path as a file object
with open(output_path, 'w') as csvfile:
    # Set the file object as a csvwriter object
    csvwriter = csv.writer(csvfile)
    # Write the list of metrics to the output file
    for metrics in report.keys():
        csvfile.write("%s,%s\n"%(metrics,report[metrics]))
```
